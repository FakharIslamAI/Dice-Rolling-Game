```python
import tkinter as tk
from tkinter import messagebox
import random
import matplotlib.pyplot as plt

class DiceGame:
    def __init__(self, root):
        self.root = root
        self.root.title("🎲 Dice Rolling Race Game")
        self.root.geometry("400x300")

        self.scores = {"Player 1": 0, "Player 2": 0}
        self.score_history = {"Player 1": [], "Player 2": []}
        self.current_player = "Player 1"

        # UI Elements
        self.title_label = tk.Label(root, text="🎲 Dice Rolling Race", font=("Arial", 16))
        self.title_label.pack(pady=10)

        self.status_label = tk.Label(root, text="Player 1's Turn", font=("Arial", 14))
        self.status_label.pack()

        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=5)

        self.score_label = tk.Label(root, text=self.get_score_text(), font=("Arial", 12))
        self.score_label.pack()

        self.roll_button = tk.Button(root, text="Roll Dice", font=("Arial", 12), command=self.play_turn)
        self.roll_button.pack(pady=10)

        self.graph_button = tk.Button(root, text="📊 Show Score Graph", font=("Arial", 12), command=self.plot_graph)
        self.graph_button.pack()

    def get_score_text(self):
        return f"Player 1: {self.scores['Player 1']} | Player 2: {self.scores['Player 2']}"

    def roll_dice(self):
        return random.randint(1, 6)

    def play_turn(self):
        roll_total = 0
        rolls = []
        while True:
            roll = self.roll_dice()
            rolls.append(roll)
            roll_total += roll
            if roll != 6:
                break

        self.scores[self.current_player] += roll_total
        self.score_history["Player 1"].append(self.scores["Player 1"])
        self.score_history["Player 2"].append(self.scores["Player 2"])

        self.result_label.config(text=f"{self.current_player} rolled {', '.join(map(str, rolls))} ➜ +{roll_total}")
        self.score_label.config(text=self.get_score_text())

        if self.scores[self.current_player] >= 100:
            messagebox.showinfo("🎉 Game Over", f"{self.current_player} wins with {self.scores[self.current_player]} points!")
            self.roll_button.config(state="disabled")
            return

        # Switch player
        if 6 not in rolls:  # if no extra roll
            self.current_player = "Player 2" if self.current_player == "Player 1" else "Player 1"
        self.status_label.config(text=f"{self.current_player}'s Turn")

    def plot_graph(self):
        turns = list(range(1, len(self.score_history["Player 1"]) + 1))

        plt.figure(figsize=(8, 5))
        plt.plot(turns, self.score_history["Player 1"], marker='o', label='Player 1')
        plt.plot(turns, self.score_history["Player 2"], marker='s', label='Player 2')
        plt.title("🎯 Score Progress Over Turns")
        plt.xlabel("Turns")
        plt.ylabel("Score")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# Run the app
if __name__ == "__main__":
    root = tk.Tk()
    app = DiceGame(root)
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import random
import matplotlib.pyplot as plt

class DiceGame:
    def __init__(self, root):
        self.root = root
        self.root.title("🎲 Dice Rolling Race Game")
        self.root.geometry("400x500")

        # 🎲 Load dice face images
        self.dice_images = {
            i: tk.PhotoImage(file=f"icons/dice{i}.png") for i in range(1, 7)
        }

        self.scores = {"Player 1": 0, "Player 2": 0}
        self.score_history = {"Player 1": [], "Player 2": []}
        self.current_player = "Player 1"

        # 🧱 UI Elements
        self.title_label = tk.Label(root, text="🎲 Dice Rolling Race", font=("Arial", 16))
        self.title_label.pack(pady=10)

        self.status_label = tk.Label(root, text="Player 1's Turn", font=("Arial", 14))
        self.status_label.pack()

        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=5)

        self.score_label = tk.Label(root, text=self.get_score_text(), font=("Arial", 12))
        self.score_label.pack()

        self.dice_label = tk.Label(root)
        self.dice_label.pack(pady=5)

        # 🎮 Buttons
        self.roll_button_p1 = tk.Button(root, text="🎯 Player 1 Roll", font=("Arial", 12), command=lambda: self.play_turn("Player 1"))
        self.roll_button_p1.pack(pady=5)

        self.roll_button_p2 = tk.Button(root, text="🎯 Player 2 Roll", font=("Arial", 12), command=lambda: self.play_turn("Player 2"))
        self.roll_button_p2.pack(pady=5)

        self.graph_button = tk.Button(root, text="📊 Show Score Graph", font=("Arial", 12), command=self.plot_graph)
        self.graph_button.pack(pady=5)

        self.restart_button = tk.Button(root, text="🔁 Restart Game", font=("Arial", 12), command=self.restart_game)
        self.restart_button.pack(pady=5)
        self.restart_button.config(state="disabled")

        self.update_buttons()

    def get_score_text(self):
        return f"Player 1: {self.scores['Player 1']} | Player 2: {self.scores['Player 2']}"

    def roll_dice(self):
        return random.randint(1, 6)

    def play_turn(self, player):
        if player != self.current_player:
            messagebox.showwarning("⚠️ Wait!", f"It's {self.current_player}'s turn!")
            return

        roll_total = 0
        rolls = []

        while True:
            roll = self.roll_dice()
            rolls.append(roll)
            roll_total += roll

            self.dice_label.config(image=self.dice_images[roll])
            self.dice_label.image = self.dice_images[roll]
            self.root.update_idletasks()

            if roll != 6:
                break

        self.scores[player] += roll_total
        self.score_history["Player 1"].append(self.scores["Player 1"])
        self.score_history["Player 2"].append(self.scores["Player 2"])

        self.result_label.config(text=f"{player} rolled {', '.join(map(str, rolls))} ➜ +{roll_total}")
        self.score_label.config(text=self.get_score_text())

        if self.scores[player] >= 100:
            messagebox.showinfo("🏆 Game Over", f"{player} wins with {self.scores[player]} points!")
            self.roll_button_p1.config(state="disabled")
            self.roll_button_p2.config(state="disabled")
            self.restart_button.config(state="normal")
            self.status_label.config(text="🎉 Game Over")
            return

        if 6 not in rolls:
            self.current_player = "Player 2" if self.current_player == "Player 1" else "Player 1"

        self.status_label.config(text=f"{self.current_player}'s Turn")
        self.update_buttons()

    def update_buttons(self):
        if self.current_player == "Player 1":
            self.roll_button_p1.config(state="normal")
            self.roll_button_p2.config(state="disabled")
        else:
            self.roll_button_p1.config(state="disabled")
            self.roll_button_p2.config(state="normal")

    def restart_game(self):
        self.scores = {"Player 1": 0, "Player 2": 0}
        self.score_history = {"Player 1": [], "Player 2": []}
        self.current_player = "Player 1"
        self.status_label.config(text="Player 1's Turn")
        self.result_label.config(text="")
        self.score_label.config(text=self.get_score_text())
        self.dice_label.config(image="")
        self.roll_button_p1.config(state="normal")
        self.roll_button_p2.config(state="disabled")
        self.restart_button.config(state="disabled")

    def plot_graph(self):
        turns = list(range(1, len(self.score_history["Player 1"]) + 1))
        plt.figure(figsize=(8, 5))
        plt.plot(turns, self.score_history["Player 1"], marker='o', label='Player 1')
        plt.plot(turns, self.score_history["Player 2"], marker='s', label='Player 2')
        plt.title("🎯 Score Progress Over Turns")
        plt.xlabel("Turns")
        plt.ylabel("Score")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()


# 🚀 Run the app
if __name__ == "__main__":
    root = tk.Tk()
    app = DiceGame(root)
    root.mainloop()
```
